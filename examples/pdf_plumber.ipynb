```python
import pdfplumber
import re

with pdfplumber.open("product_specs_amazon_style.pdf") as pdf:
    for page in pdf.pages:
        print(page.extract_text())
        text = page.extract_text()

        pattern = r'([^\n]+)\nBrand:'
        headings = re.findall(pattern, text)
        print('__________________')
        print(headings)



```

```python
from langchain_community.document_loaders import PDFPlumberLoader

loader = PDFPlumberLoader("product_specs_amazon_style.pdf")
```

```python
docs = loader.load()
```

```python
docs
```

```python
print(docs[0].metadata)
```

```python
text = "\n".join([doc.page_content for doc in docs])

text
```

```python
import re

text = re.sub(r'\(cid:\d+\)', '', text)
text = re.sub(r'\n+', '\n', text)  # normalize line breaks
```

```python
text
```

```python
pattern = r'([^\n]+\nBrand:.*?)(?=\n[^\n]+\nBrand:|\Z)'

products = re.findall(pattern, text, re.DOTALL)

for i, p in enumerate(products, 1):
    print(f"\n--- Product {i} ---\n")
    print(p)
```
